```python
# ==========================================
# Week 3  Hypotheses, Analysis Plan & Risks
# Dataset: city_day.csv
# Goal: Validate hypothesis directions using EDA + simple statistical checks
# ==========================================

# -------------------------
# 0. Imports
# -------------------------
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

# Use seaborn default style for consistent plots
sns.set()

# -------------------------
# 3. Basic cleaning: dates, seasons
# -------------------------
# Ensure Date is datetime for extracting time-based features (Year/Month/Season)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Remove rows with invalid/unparseable dates (cannot be used for seasonal analysis)
df = df.dropna(subset=["Date"])

# Time features used for seasonal / trend-based comparisons
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month

# Map month -> season to support H2 comparisons
def month_to_season(m):
    if m in [12, 1, 2]:
        return "Winter"
    elif m in [3, 4, 5]:
        return "Spring"
    elif m in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["Season"] = df["Month"].apply(month_to_season)

print("\nColumns after adding Year, Month, Season:")
print(df.columns.tolist())


# -------------------------
# 4. Missing values (for H4 + risk analysis)
# -------------------------
# Summarize missingness to identify whether it may be city/time dependent
print("\nMissing values per column:")
missing_summary = df.isnull().sum().sort_values(ascending=False)
print(missing_summary)

# Missing percentage gives an interpretable scale for reporting
missing_pct = (df.isnull().mean() * 100).round(2)
print("\nPercentage of missing values per column:")
print(missing_pct)


# -------------------------
# 5. Correlation with AQI (H1 & part of H3)
# -------------------------
# Select pollutant columns only if they exist in the dataset (robust to schema differences)
pollutant_cols = [
    col for col in ["PM2.5", "PM10", "NO2", "SO2", "CO", "O3", "NOx", "NH3"]
    if col in df.columns
]

# Target variable for the analysis
target_col = "AQI"

# Correlation requires AQI present; remove missing AQI rows for correlation analysis
df_corr = df.dropna(subset=[target_col])

# Compute correlations among pollutants and AQI
corr_matrix = df_corr[pollutant_cols + [target_col]].corr()

print("\nCorrelation matrix (pollutants vs AQI):")
display(corr_matrix[target_col].sort_values(ascending=False))

# Heatmap for quick visual comparison of relationships
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f")
plt.title("Correlation Matrix: Pollutants and AQI")
plt.tight_layout()
plt.show()

# Scatter plots for key particulate pollutants
for col in ["PM2.5", "PM10"]:
    if col in df_corr.columns:
        plt.figure(figsize=(6, 4))
        sns.scatterplot(data=df_corr, x=col, y=target_col, alpha=0.4)
        plt.title(f"{col} vs AQI")
        plt.tight_layout()
        plt.show()


# -------------------------
# 6. Seasonal analysis 
# -------------------------
if target_col in df.columns:
    # Seasonal mean levels to check winter vs summer differences
    seasonal_means = df.groupby("Season")[pollutant_cols + [target_col]].mean()

    print("\nSeasonal mean pollutant levels and AQI:")
    display(seasonal_means)

    # Distribution comparison of AQI across seasons
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df, x="Season", y=target_col, order=["Winter", "Spring", "Summer", "Autumn"])
    plt.title("AQI Distribution by Season")
    plt.tight_layout()
    plt.show()

    # One-way ANOVA: tests whether mean AQI differs significantly across seasons
    groups = []
    for season in ["Winter", "Spring", "Summer", "Autumn"]:
        group_vals = df.loc[df["Season"] == season, target_col].dropna()
        if len(group_vals) > 0:
            groups.append(group_vals)

    if len(groups) >= 2:
        f_stat, p_val = stats.f_oneway(*groups)
        print(f"\nANOVA for AQI across seasons  F-statistic: {f_stat:.3f}, p-value: {p_val:.5f}")
    else:
        print("\nNot enough seasonal groups for ANOVA.")


# -------------------------
# 7. City-wise correlation patterns (H3)
# -------------------------
if "City" in df.columns and target_col in df.columns:
    # Compute pollutantAQI correlation separately per city to see variation in dominant drivers
    city_corr_list = []

    for city, sub in df.groupby("City"):
        sub = sub.dropna(subset=[target_col])

        # Skip cities with low sample size (unstable correlations)
        if len(sub) < 30:
            continue

        for pol in pollutant_cols:
            # Require enough valid pollutant observations for correlation to be meaningful
            if pol in sub.columns and sub[pol].notnull().sum() > 10:
                r = sub[[pol, target_col]].corr().iloc[0, 1]
                city_corr_list.append({"City": city, "Pollutant": pol, "Corr_with_AQI": r})

    city_corr_df = pd.DataFrame(city_corr_list)
    print("\nExample of city-wise pollutantAQI correlations (first 10 rows):")
    display(city_corr_df.head(10))


# -------------------------
# 8. Missingness pattern 
# -------------------------
if "City" in df.columns:
    # Compare missing AQI rates across cities to check non-random missingness
    if target_col in df.columns:
        city_missing_aqi = (
            df.groupby("City")[target_col]
              .apply(lambda x: x.isnull().mean())
              .sort_values(ascending=False)
        )

        print("\nFraction of missing AQI values per city (top 10):")
        display(city_missing_aqi.head(10))

    # Visual missingness check (random sample to keep plot readable)
    sample_df = df.sample(min(500, len(df)), random_state=42)
    plt.figure(figsize=(10, 4))
    sns.heatmap(sample_df[pollutant_cols + [target_col]].isnull(), cbar=False)
    plt.title("Missingness Pattern (Sample of 500 Rows)")
    plt.xlabel("Variables")
    plt.ylabel("Sampled Rows")
    plt.tight_layout()
    plt.show()


# -------------------------
# 9. Anticipated results & risks 
# -------------------------
# Summarize expected trends and key data risks identified in Week 3
print("\n--- Anticipated Results---")
print("""
- PM2.5 and PM10 are expected to show strong positive correlation with AQI.
- Winter is expected to have higher average AQI and pollutant levels than Summer.
- Different cities are expected to show different dominant pollutants.
- Missing data may be concentrated in specific cities or periods.
""")

print("\n--- Potential Risks  ---")
print("""
- High proportion of missing values in some pollutants or cities.
- Uneven data coverage across cities, leading to bias.
- Outliers in pollutant levels affecting correlations and tests.
- Possible non-linear relationships that simple correlation cannot capture.
""")
```

```python
# ==========================================
# Weeks 46: Process the data and extract useful features.
# ==========================================

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder  # (not used below, safe to remove if not needed)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# -------------------------
# 1. Basic cleaning + time feature creation
# -------------------------

# Parse Date into datetime so we can extract time-based predictors
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Drop rows with invalid dates 
df = df.dropna(subset=["Date"])

# Time features help the model capture trends / seasonality patterns
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Week"] = df["Date"].dt.isocalendar().week

# Map month -> season (high-level seasonal signal for AQI/pollution)
def month_to_season(m):
    if m in [12, 1, 2]:
        return "Winter"
    elif m in [3, 4, 5]:
        return "Spring"
    elif m in [6, 7, 8]:
        return "Summer"
    else:
        return "Autumn"

df["Season"] = df["Month"].apply(month_to_season)

print("New columns added: Year, Month, Day, Week, Season")


# -------------------------
# 2. Select pollutant predictors + target
# -------------------------
# Keep only pollutant columns that exist in the dataset (robust to schema differences)
pollutant_cols = [
    col for col in ["PM2.5", "PM10", "NO2", "NOx", "SO2", "CO", "O3", "NH3"]
    if col in df.columns
]

# Target variable to predict
target_col = "AQI"


# -------------------------
# 3. Handle missing values
# -------------------------

# Impute missing pollutant values using median (robust to skew/outliers)
num_imputer = SimpleImputer(strategy="median")
df[pollutant_cols] = num_imputer.fit_transform(df[pollutant_cols])

# AQI is the label; rows without AQI cannot be used for supervised training
df = df.dropna(subset=[target_col])

# Ensure Season has no missing values (required before encoding)
cat_imputer = SimpleImputer(strategy="most_frequent")
df[["Season"]] = cat_imputer.fit_transform(df[["Season"]])

print("Missing values processed.")


# -------------------------
# 4. Outlier treatment (IQR clipping)
# -------------------------
# Clip extreme pollutant values to reduce undue influence of rare spikes
for col in pollutant_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Winsorization-style clipping (keeps all rows, limits extremes)
    df[col] = np.where(
        df[col] < lower, lower,
        np.where(df[col] > upper, upper, df[col])
    )

print("Outliers treated using IQR method.")


# -------------------------
# 5. Feature engineering
# -------------------------

# Sort by City and Date before rolling statistics to ensure correct window order
df = df.sort_values(["City", "Date"])

# Rolling averages capture short-term and weekly smoothed pollution dynamics
for col in pollutant_cols:
    df[f"{col}_rolling3"] = df.groupby("City")[col].transform(lambda x: x.rolling(3).mean())
    df[f"{col}_rolling7"] = df.groupby("City")[col].transform(lambda x: x.rolling(7).mean())

# Aggregate pollution index (single combined indicator across pollutants)
df["Pollution_Load"] = df[pollutant_cols].sum(axis=1)

# One-hot encode Season to represent categorical seasonal effects numerically
df = pd.get_dummies(df, columns=["Season"], drop_first=True)

print("Feature engineering completed (rolling averages, pollution load, season encoding).")


# -------------------------
# 6. Numerical scaling
# -------------------------
# Standardize continuous features for comparability and model stability (esp. linear models)
scaler = StandardScaler()

# NOTE: currently only scaling raw pollutants + Pollution_Load (rolling features remain unscaled)
scaled_cols = pollutant_cols + ["Pollution_Load"]
df[scaled_cols] = scaler.fit_transform(df[scaled_cols])

print("Numerical features scaled using StandardScaler.")


# -------------------------
# 7. Final feature matrix (X) and target vector (y)
# -------------------------
# Define final set of predictors used for modeling
feature_cols = (
    pollutant_cols
    + [f"{col}_rolling3" for col in pollutant_cols]
    + [f"{col}_rolling7" for col in pollutant_cols]
    + ["Pollution_Load"]
    + [col for col in df.columns if col.startswith("Season_")]
    + ["Year", "Month", "Week"]
)

X = df[feature_cols]
y = df[target_col]

print("Final feature matrix (X) and target vector (y) prepared.")
print("X shape:", X.shape)
print("y shape:", y.shape)

X.head()


```

```python
# ==========================================
# Weeks 69: Build and Evaluate Models
# Task: Predict AQI from pollutant & engineered features
# ==========================================

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# -------------------------
# 0. Remove NaNs from modeling dataset
# -------------------------
# Rolling features and other engineered features may introduce NaNs.
# This step ensures training/evaluation uses complete rows only.

print("Original X shape:", X.shape)
print("Original y shape:", y.shape)

# Join features + target to remove any row with missing values consistently
data_for_model = pd.concat([X, y], axis=1)

print("Total missing values before cleaning:", data_for_model.isnull().sum().sum())

# Drop rows with missing values in either X or y
data_for_model = data_for_model.dropna()

# Split back into clean feature matrix and target vector
X_clean = data_for_model[X.columns]
y_clean = data_for_model[y.name]  # expected: 'AQI'

print("Cleaned X shape:", X_clean.shape)
print("Cleaned y shape:", y_clean.shape)
print("Total missing values after cleaning:",
      pd.concat([X_clean, y_clean], axis=1).isnull().sum().sum())


# -------------------------
# 1. Traintest split
# -------------------------
# Hold-out test set provides unbiased evaluation of final model performance.
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y_clean,
    test_size=0.2,
    random_state=42
)

print("\nTrain shape:", X_train.shape, " Test shape:", X_test.shape)


# -------------------------
# 2. Helper: model training + evaluation
# -------------------------
# Standardizes evaluation across models:
# - Train/Test MAE and RMSE (error magnitude)
# - Train/Test R (explained variance)
# - Cross-validated R on training set (robustness / stability)

def evaluate_regression_model(model, X_train, y_train, X_test, y_test, cv_folds=5):
    """
    Fit the model, evaluate on train & test sets,
    and run k-fold CV (R) on the training set.
    """

    # Train model
    model.fit(X_train, y_train)

    # Predictions on train and test
    y_pred_train = model.predict(X_train)
    y_pred_test = model.predict(X_test)

    # MAE: average absolute prediction error (same units as AQI)
    mae_train = mean_absolute_error(y_train, y_pred_train)
    mae_test = mean_absolute_error(y_test, y_pred_test)

    # RMSE: penalizes larger errors more strongly than MAE
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

    # R: goodness-of-fit (higher is better)
    r2_train = r2_score(y_train, y_pred_train)
    r2_test = r2_score(y_test, y_pred_test)

    # Cross-validation (training only) to estimate generalization stability
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv_folds, scoring="r2")

    print("Model:", model.__class__.__name__)
    print("-------------------------------------")
    print(f"Train MAE:  {mae_train:.3f}")
    print(f"Test MAE:   {mae_test:.3f}")
    print(f"Train RMSE: {rmse_train:.3f}")
    print(f"Test RMSE:  {rmse_test:.3f}")
    print(f"Train R:   {r2_train:.3f}")
    print(f"Test R:    {r2_test:.3f}")
    print(f"{cv_folds}-fold CV R: mean={cv_scores.mean():.3f}, std={cv_scores.std():.3f}")
    print()

    return {
        "model": model,
        "mae_train": mae_train,
        "mae_test": mae_test,
        "rmse_train": rmse_train,
        "rmse_test": rmse_test,
        "r2_train": r2_train,
        "r2_test": r2_test,
        "cv_scores": cv_scores
    }


# -------------------------
# 3. Baseline model
# -------------------------
# DummyRegressor provides a minimum benchmark (predicts mean AQI for all cases).
baseline = DummyRegressor(strategy="mean")
baseline_results = evaluate_regression_model(baseline, X_train, y_train, X_test, y_test)


# -------------------------
# 4. Linear regression (simple baseline with features)
# -------------------------
lin_reg = LinearRegression()
lin_results = evaluate_regression_model(lin_reg, X_train, y_train, X_test, y_test)


# -------------------------
# 5. Ridge regression (regularized linear model)
# -------------------------
# Ridge can reduce overfitting / multicollinearity effects via L2 regularization.
ridge = Ridge(alpha=1.0, random_state=42)
ridge_results = evaluate_regression_model(ridge, X_train, y_train, X_test, y_test)


# -------------------------
# 6. Random Forest regressor
# -------------------------
# Tree ensemble capturing nonlinear relationships and feature interactions.
rf = RandomForestRegressor(
    n_estimators=200,
    max_depth=None,
    random_state=42,
    n_jobs=-1
)
rf_results = evaluate_regression_model(rf, X_train, y_train, X_test, y_test)


# -------------------------
# 7. Gradient Boosting regressor
# -------------------------
# Boosting often performs well on structured tabular data with nonlinearities.
gbr = GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)
gbr_results = evaluate_regression_model(gbr, X_train, y_train, X_test, y_test)


# -------------------------
# 8. Compare model performance on test set
# -------------------------
# Summary table helps select the best-performing model for reporting.
summary = pd.DataFrame({
    "Model": [
        "Baseline (Mean)",
        "Linear Regression",
        "Ridge Regression",
        "Random Forest",
        "Gradient Boosting"
    ],
    "Test MAE": [
        baseline_results["mae_test"],
        lin_results["mae_test"],
        ridge_results["mae_test"],
        rf_results["mae_test"],
        gbr_results["mae_test"],
    ],
    "Test RMSE": [
        baseline_results["rmse_test"],
        lin_results["rmse_test"],
        ridge_results["rmse_test"],
        rf_results["rmse_test"],
        gbr_results["rmse_test"],
    ],
    "Test R2": [
        baseline_results["r2_test"],
        lin_results["r2_test"],
        ridge_results["r2_test"],
        rf_results["r2_test"],
        gbr_results["r2_test"],
    ]
})

print("\n=== Model Comparison on Test Set ===")
display(summary.sort_values(by="Test R2", ascending=False))


# -------------------------
# 9. Residual diagnostics for selected best model
# -------------------------
# Residual plot checks systematic errors (e.g., bias at high AQI values).
best_model = rf  # update if another model has the best Test R / lowest errors

y_pred_test_best = best_model.predict(X_test)
residuals = y_test - y_pred_test_best

plt.figure(figsize=(6, 4))
sns.scatterplot(x=y_pred_test_best, y=residuals, alpha=0.4)
plt.axhline(0, linestyle="--")
plt.xlabel("Predicted AQI")
plt.ylabel("Residuals (y_true - y_pred)")
plt.title(f"Residual Plot  {best_model.__class__.__name__}")
plt.tight_layout()
plt.show()


# -------------------------
# 10. Feature importance (tree-based models only)
# -------------------------
# For tree models, feature_importances_ provides a ranking of influential predictors.
if hasattr(best_model, "feature_importances_"):
    importances = best_model.feature_importances_
    feat_importance_df = pd.DataFrame({
        "Feature": X_clean.columns,
        "Importance": importances
    }).sort_values(by="Importance", ascending=False)

    print("\nTop 15 features by importance:")
    display(feat_importance_df.head(15))

    plt.figure(figsize=(8, 6))
    sns.barplot(
        data=feat_importance_df.head(15),
        x="Importance",
        y="Feature"
    )
    plt.title(f"Top 15 Feature Importances  {best_model.__class__.__name__}")
    plt.tight_layout()
    plt.show()
```
